```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt

# Chargement des donnÃ©es
df = pd.read_csv('Data.csv', sep=";")  # Assurez-vous que le chemin est correct
# Filtrer les donnÃ©es pour l'entreprise 'CS.PA'
df_filtered = df[df['Symbole'] == 'CS.PA']

# PrÃ©paration des donnÃ©es
df_filtered['Date'] = pd.to_datetime(df_filtered['Date'], format='%Y-%m-%d')
df_filtered.sort_values('Date', inplace=True)

# SÃ©lection des caractÃ©ristiques et de la cible
X = df_filtered[["Date"]]  # Variables indÃ©pendantes
y = df_filtered['Close']  # Variable Ã  prÃ©dire

# Division des donnÃ©es en un ensemble d'entraÃ®nement et un ensemble de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# EntraÃ®nement du modÃ¨le
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# PrÃ©diction sur l'ensemble de test
y_pred = regressor.predict(X_test)

# Ã‰valuation du modÃ¨le
print('Erreur quadratique moyenne (RMSE):', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))

# CrÃ©ation d'un DataFrame pour les jours futurs
future_days = pd.DataFrame(np.zeros((60, 4)), columns=X.columns)

# PrÃ©diction pour les 60 prochains jours
future_pred = regressor.predict(future_days)

# CrÃ©ation du graphique
plt.figure(figsize=(10, 6))
plt.plot(df_filtered["Date"], df_filtered['Close'], label='DonnÃ©es historiques')
plt.plot(pd.date_range(start=df_filtered["Date"].max(), periods=60), future_pred, label='PrÃ©dictions pour les 60 prochains jours', color='red')
plt.xlabel('Jours')
plt.ylabel('Valeur prÃ©dite')
plt.title('PrÃ©dictions de rÃ©gression linÃ©aire pour les 60 prochains jours')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.tree import DecisionTreeRegressor  # Importer le modÃ¨le de rÃ©gression par arbre de dÃ©cision
from sklearn import metrics
import numpy as np
from pandas.tseries.offsets import BDay
from scipy import stats
import matplotlib.pyplot as plt

# Charger les donnÃ©es
df = pd.read_csv('fichiers/Data.csv',sep=";")
df = df[df['Symbole'] == 'KER.PA']
# Convertir la colonne de date en datetime et extraire le nombre de jours
df['Date'] = pd.to_datetime(df['Date'])
start_date = pd.to_datetime('2024-02-01')
df = df[df['Date'] >= start_date]
df['Date'] = (df['Date'] - df['Date'].min())  / np.timedelta64(1,'D')

# DÃ©finir X et y
X = df['Date'].values.reshape(-1,1)
y = df['Close'].values.reshape(-1,1)

# Diviser les donnÃ©es en ensemble d'entraÃ®nement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# EntraÃ®ner le modÃ¨le
regressor = DecisionTreeRegressor()  # Utiliser le modÃ¨le de rÃ©gression par arbre de dÃ©cision
regressor.fit(X_train, y_train)

# Le reste du code reste le mÃªme
# DÃ©finir la date de dÃ©but pour les prÃ©visions, en excluant les week-ends
start_forecast_date = pd.to_datetime('2024-03-11')

# Calculer le nombre de jours ouvrables Ã  prÃ©dire
days_in_future = 30  # Le nombre total de jours Ã  prÃ©dire
future_business_days = pd.date_range(start_forecast_date, periods=days_in_future, freq=BDay())

# Convertir les dates futures en nombres, basÃ© sur le calcul original
future_forecast_days = [(day - (start_date + pd.Timedelta(days=df['Date'].min()))).days for day in future_business_days]
future_forecast = np.array(future_forecast_days).reshape(-1,1)

# PrÃ©diction avec le modÃ¨le entraÃ®nÃ©
y_pred = regressor.predict(future_forecast)

# CrÃ©ation d'un DataFrame pour les prÃ©dictions futures
future_predictions_df = pd.DataFrame({
    'Future Date': future_business_days,
    'Predicted Close': y_pred.flatten()
})

# Afficher les prÃ©dictions
print(future_predictions_df.head(10))



residuals = y_test - regressor.predict(X_test)

# Calculer l'erreur standard des rÃ©sidus
stderr = residuals.std()

# Nombre de prÃ©dictions
n = future_forecast.shape[0]

# DegrÃ©s de libertÃ©
df = len(X_test) - 2

# Trouver le t-score pour un intervalle de confiance de 95%
t_score = stats.t.ppf(0.975, df)

# Calculer l'intervalle de confiance pour chaque prÃ©diction
confidence_interval = t_score * stderr / np.sqrt(n)

# Ajouter l'intervalle de confiance aux prÃ©dictions
future_predictions_df['Lower Bound'] = future_predictions_df['Predicted Close'] - confidence_interval
future_predictions_df['Upper Bound'] = future_predictions_df['Predicted Close'] + confidence_interval

# Afficher les prÃ©dictions avec l'intervalle de confiance
print(future_predictions_df.head(10))

# Sauvegarde du DataFrame dans un fichier CSV
future_predictions_path = 'future_predictions.csv'  # Adjusted path for saving in accessible location
future_predictions_df.to_csv(future_predictions_path, index=False)

lower_bound = future_predictions_df['Lower Bound']
upper_bound = future_predictions_df['Upper Bound']
# CrÃ©ation du graphique
plt.figure(figsize=(10, 6))
plt.plot(future_business_days, y_pred, color='blue', label='PrÃ©dictions linÃ©aires')
plt.fill_between(future_business_days, lower_bound, upper_bound, color='gray', alpha=0.2, label='Intervalle de confiance 95%')
plt.xlabel('Date')
plt.ylabel('PrÃ©diction Close')
plt.title('PrÃ©dictions avec intervalle de confiance Ã  95%')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, learning_curve
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
import matplotlib.pyplot as plt
import numpy as np
from pandas.tseries.offsets import BDay

# Fonction pour crÃ©er le modÃ¨le, nÃ©cessaire pour KerasRegressor
def create_model():
    model = Sequential()
    model.add(Dense(50, input_dim=1, activation='relu'))
    model.add(Dense(30, activation='relu'))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model

# Charger les donnÃ©es
df = pd.read_csv('Data.csv',sep=";")
df = df[df['Symbole'] == 'KER.PA']
# Convertir la colonne de date en datetime et extraire le nombre de jours
df['Date'] = pd.to_datetime(df['Date'])
start_date = pd.to_datetime('2024-02-01')
df = df[df['Date'] >= start_date]
df['Date'] = (df['Date'] - df['Date'].min())  / np.timedelta64(1,'D')

# DÃ©finir X et y
X = df['Date'].values.reshape(-1,1)
y = df['Close'].values.reshape(-1,1)

# Diviser les donnÃ©es en ensemble d'entraÃ®nement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# CrÃ©er le modÃ¨le
model = KerasRegressor(build_fn=create_model, epochs=50, batch_size=10, verbose=0)

# EntraÃ®ner le modÃ¨le
model.fit(X_train, y_train)

# Fonction pour tracer les courbes d'apprentissage
def plot_learning_curve(model, X, y):
    train_sizes, train_scores, test_scores = learning_curve(model, X, y, cv=5)

    train_scores_mean = np.mean(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)

    plt.plot(train_sizes, train_scores_mean, label='Score d\'entraÃ®nement')
    plt.plot(train_sizes, test_scores_mean, label='Score de validation')

    plt.ylabel('Score')
    plt.xlabel('Taille de l\'Ã©chantillon d\'entraÃ®nement')
    plt.title('Courbes d\'apprentissage')
    plt.legend(loc='best')
    plt.show()

# Utilisez la fonction sur votre modÃ¨le
plot_learning_curve(model, X, y)
```
